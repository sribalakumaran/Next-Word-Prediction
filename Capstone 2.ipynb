```python
import torch
from transformers import GPT2Tokenizer, GPT2Model
from typing import List, Tuple
import nltk
from nltk.tokenize import word_tokenize

# Initialize tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
tokenizer.pad_token = tokenizer.eos_token
model = GPT2Model.from_pretrained('gpt2')
model.eval()
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

def sentences_to_gpt2_embeddings(sentences: List[str]) -> torch.Tensor:
    inputs = tokenizer(sentences, return_tensors='pt', padding=True, truncation=True)
    inputs = {key: val.to(device) for key, val in inputs.items()}
    with torch.no_grad():
        outputs = model(**inputs)
        last_hidden_states = outputs.last_hidden_state
    attention_mask = inputs['attention_mask']
    masked_hidden = last_hidden_states * attention_mask.unsqueeze(-1)
    embeddings = masked_hidden.sum(dim=1) / attention_mask.sum(dim=1, keepdim=True)
    return embeddings.cpu()

def prepare_data(sentences: List[str]) -> Tuple[List[torch.Tensor], List[str]]:
    embeddings = []
    targets = []

    for sentence in sentences:
        words = word_tokenize(sentence.lower())
        if len(words) < 2:  # Skip sentences with insufficient words
            continue
        # Use all but the last word as input, predict the last word
        input_sentence = ' '.join(words[:-1])
        target_word = words[-1]

        embedding = sentences_to_gpt2_embeddings([input_sentence])[0]  # Get embedding for input
        embeddings.append(embedding)
        targets.append(target_word)

    return embeddings, targets

sentences = [
    "The quick brown fox jumps",
    "A cat sits on the",
    "Sunny hills bloom with vivid"
]

embeddings, targets = prepare_data(sentences)
print(f"Embeddings shape: {embeddings[0].shape}")
print(f"Targets: {targets}")
```

```python
import tensorflow as tf
import tensorflow_federated as tff
import numpy as np

def create_federated_dataset(embeddings: List[torch.Tensor], targets: List[str], num_clients: int = 3):
    vocab = list(set(targets))
    word_to_idx = {word: idx for idx, word in enumerate(vocab)}
    target_indices = [word_to_idx[word] for word in targets]

    embeddings_np = [emb.numpy() for emb in embeddings]

    dataset = tf.data.Dataset.from_tensor_slices({
        'embeddings': embeddings_np,
        'targets': target_indices
    })

    dataset = dataset.shuffle(len(embeddings)).batch(1)

    client_datasets = []
    dataset_list = list(dataset)
    client_size = len(dataset_list) // num_clients

    for i in range(num_clients):
        start = i * client_size
        end = (i + 1) * client_size if i < num_clients - 1 else len(dataset_list)
        client_data = tf.data.Dataset.from_tensor_slices({
            'embeddings': [dataset_list[j]['embeddings'].numpy() for j in range(start, end)],
            'targets': [dataset_list[j]['targets'].numpy() for j in range(start, end)]
        }).batch(1)
        client_datasets.append(client_data)

    return client_datasets, vocab

client_datasets, vocab = create_federated_dataset(embeddings, targets)
print(f"Number of clients: {len(client_datasets)}")
print(f"Vocabulary: {vocab}")
```

```python
def create_keras_model(vocab_size: int, embedding_dim: int = 768):
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(embedding_dim,)),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(vocab_size, activation='softmax')
    ])
    return model

def model_fn():
    keras_model = create_keras_model(vocab_size=len(vocab))
    return tff.learning.models.from_keras_model(
        keras_model,
        input_spec=client_datasets[0].element_spec,
        loss=tf.keras.losses.SparseCategoricalCrossentropy(),
        metrics=[tf.keras.metrics.SparseCategoricalAccuracy()]
    )
```

```python
# Define the federated learning process
iterative_process = tff.learning.algorithms.build_weighted_fed_avg(
    model_fn,
    client_optimizer_fn=lambda: tf.keras.optimizers.SGD(0.1),
    server_optimizer_fn=lambda: tf.keras.optimizers.SGD(1.0)
)

# Initialize the server state
state = iterative_process.initialize()

# Train for a few rounds
num_rounds = 5
for round_num in range(1, num_rounds + 1):
    state, metrics = iterative_process.next(state, client_datasets)
    print(f'Round {round_num}, Metrics: {metrics}')
```

```python
def predict_next_word(sentence: str, trained_model, vocab: List[str]):
    """Predict the next word for a given sentence."""
    embedding = sentences_to_gpt2_embeddings([sentence])[0].numpy()
    embedding = tf.expand_dims(embedding, axis=0)  # Add batch dimension
    logits = trained_model.predict(embedding)
    predicted_idx = tf.argmax(logits, axis=1).numpy()[0]
    return vocab[predicted_idx]

trained_keras_model = create_keras_model(vocab_size=len(vocab))

test_sentence = "The quick brown fox"
predicted_word = predict_next_word(test_sentence, trained_keras_model, vocab)
print(f"Input: {test_sentence}")
print(f"Predicted next word: {predicted_word}")
```
